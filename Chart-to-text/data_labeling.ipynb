```python
import os
import pandas as pd
```

```python
labeling_path = '/content/drive/MyDrive/chart_generating/chart_generating/Chart2text_datasets_6050/Chart2Text_labeling.xlsx'
labeling = pd.read_excel(labeling_path)
data_path = '/content/drive/MyDrive/chart_generating/chart_generating/Chart2text_datasets_6050/data_ko/' #경로 argparse로 받을 수 있도록 수정 예정
csv_list = [file for file in os.listdir(data_path)]
txt_path = '/content/drive/MyDrive/chart_generating/chart_generating/Chart2text_datasets_6050/tit_ko/'
sorted_csv_list = sorted(csv_list, key=lambda x: int(x.split('.')[0])) #순서대로 정렬
labeling['combined'] = labeling['file name'].astype(str) + '_' + labeling['type'].astype(str) + '_' + labeling['axis'].astype(str) #라벨링된 항목들을 병합
labeling_list = labeling['combined'].tolist()


for csv_name in sorted_csv_list:
  num = int(os.path.splitext(csv_name)[0])-1
  new_name = labeling_list[num] #라벨링된 파일명으로 변경
  old_csv = os.path.join(data_path, csv_name) # 경로 지정
  labeled_csv = os.path.join(data_path, new_name+'.csv')
  #print(old_path, new_path)
  os.rename(old_csv, labeled_csv)

  txt_file = txt_path+csv_name
  #txt파일 라벨링
  txt_name = os.path.splitext(txt_file)[0]+'.txt'
  old_txt = os.path.join(txt_path, txt_name)
  labeled_txt = os.path.join(txt_path, new_name+'.txt')
  os.rename(old_txt, labeled_txt)
print('finished')
```
